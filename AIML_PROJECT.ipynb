# LOGISTIC REGRESSION

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score, confusion_matrix
```

```python
# Load dataset
df = pd.read_csv('/content/drive/MyDrive/archive.zip')
```

```python
# Define target variable
target = 'Sleep_Quality_Score'
```

```python
# Define features
features = ['Heart_Rate_Variability', 'Body_Temperature', 'Bedtime_Consistency']
```

```python
# Convert target variable into numerical values (0/1) using LabelEncoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df[target] = le.fit_transform(df[target])
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)
```

```python
# Create logistic regression model
model = LogisticRegression(max_iter=1000)
```

```python
# Train model
model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(X_test)
```

```python
# Calculate metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
conf_mat = confusion_matrix(y_test, y_pred)
```

```python
# Print metrics
print("Accuracy:", accuracy)
print("Precision:", precision)
print("F1 Score:", f1)
print("Recall:", recall)
print("Confusion Matrix:\n", conf_mat)
```

# ROC CURVE

Step 1: Import necessary libraries and load the dataset

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv('/content/drive/MyDrive/archive.zip')
```

```python
# Define target variable
target = 'Sleep_Quality_Score'
```

```python
# Convert target variable to binary (0/1)
df[target] = df[target].apply(lambda x: 1 if x >= 4 else 0)
```

Step 2: Split data into training and testing sets

```python
# Define features
features = ['Heart_Rate_Variability', 'Body_Temperature', 'Bedtime_Consistency']
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)
```

Step 3: Train a classifier (Random Forest)

```python
# Create Random Forest classifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
# Train model
model.fit(X_train, y_train)
```

Step 4: Predict probabilities and plot ROC curve

```python
# Predict probabilities
y_pred_proba = model.predict_proba(X_test)[:, 1]
```

```python
# Plot ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
plt.plot(fpr, tpr, color='blue', label='ROC curve')
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

```python
# Calculate AUC-ROC score
auc = roc_auc_score(y_test, y_pred_proba)
print("AUC-ROC score:", auc)
```
